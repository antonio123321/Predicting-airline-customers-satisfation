## Let' s have another test to check if the data is normaly distributed:
#### I could use shapiro.test but the data points are more then 5000 so this kind of test will be no accurate
#### I can use instead: Kolmogorov-Smirnov Test:

```python
from scipy.stats import kstest

# Run the Kolmogorov-Smirnov test
res_for_normality = kstest(data['Arrival Delay'], 'norm')

#Check the p-value
if res_for_normality.pvalue > 0.05:
    print("Data appears to be normally distributed")
else:
    print("Data does not appear to be normally distributed")
```

```python

import statsmodels.api as sm

# Create a Q-Q plot
sm.qqplot(d, line='q', fit=True, marker='o', color='blue', markersize=2)
plt.xlabel('Theoretical')
plt.ylabel('Sample')
plt.show()
```

```python
X=data_no_missing_val.iloc[:,:-1].values
y=data_no_missing_val.iloc[:,-1].values
y  # y has one dim which is the format we need to keep it.
#X will have 2 dimension . again the format we need to have it
```

# now I split the data to train and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,stratify=y,random_state=17)
print(len(data_no_missing_val))
print(len(X_train))
print(len(X_test))
```

# Feature selection:


```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=0)# this use Random forest as the base model
rfecv=RFECV(rf,cv=3,step=2)#You create an instance of RFECV, specifying the base model (Random Forest Classifier),
                            #the number of cross-validation folds (cv=3), and the step for feature removal (step=5)

X_train_fs=rfecv.fit_transform(X_train,y_train)
X_test_fs=rfecv.transform(X_test)

print(f'Number of remaining features: {X_train_fs.shape[1]}')
```

```python
X_train_fs=pd.DataFrame(X_train_fs)
X_test_fs=pd.DataFrame(X_test_fs)
X_test_fs
```

```python
X_test_fs.iloc[:,3:5]
```

```python
from imblearn.over_sampling import SMOTE

resampler=SMOTE(random_state=0)
X_train_fs_smote,y_train_smote=resampler.fit_resample(X_train_fs,y_train)

sns.countplot(x=y_train_smote)
```

# now we need to scale the data, notice that X_train have to have fit_trasform while X_test need only transform. 

```python
# scale all data which is not 0 or 1
# I could avoid this step as decision tree is robust to unscaled data


from sklearn.preprocessing import MinMaxScaler      #I use min-max scaler so all of the values will be between 1 and  0

sc=MinMaxScaler()
X_train_fs_smote=sc.fit_transform(X_train_fs_smote) # here you need to scale all the column with numerical values that are not 0 or one's. meaning from Age(index 4) forward
X_test_fs=sc.transform(X_test_fs)
X_train[0]
```

## now the data preprocessin is ended

# create and train the model:


```python
model=tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(128,activation='relu',input_shape=(18,)))#18 is the number of input layers in the data (columns features)
model.add(tf.keras.layers.Dense(128,activation='relu'))

model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

now compile the model:

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # if we were predicting a binary output  in loss put:  binary_crossentropy
              metrics='accuracy')
```

```python
model.summary()
```

train the model:

```python
history=model.fit(X_train_fs_smote,y_train_smote,batch_size=16,epochs=15,verbose=2,validation_split=0.3)
```

# predict y_pred based on X_train so that later you can evaluate the model comparing y_pred to y_true(y_test)

```python
#THIS GIVE WRONG RESULTS WITH ALL PREDICTIONS=0

# y_pred=model.predict(X_test_fs)
# y_pred=y_pred.argmax(axis=1)
# y_pred
```

```python
y_pred = model.predict(X_test_fs)
y_pred = (y_pred > 0.5).astype(int)
y_pred
```

# create a dataframe to compare the 2 variables:

```python
comparison=pd.DataFrame(y_test,columns=['y_true'])
comparison['y_pred']=y_pred
comparison
```

# evaluate the model:

```python
from sklearn import metrics
accur=metrics.accuracy_score(y_test,y_pred)
print('accuracy is: ',accur)
```

```python
accuracy=history.history['accuracy']
validation_accuracy=history.history['val_accuracy']

plt.plot(accuracy,label='Training Set Accuracy')
plt.plot(validation_accuracy,label='Validation Set Accuracy ')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and validation Accuracy Across eposchs')
plt.legend()
```

```python
loss=history.history['loss']
validation_loss=history.history['val_loss']

plt.plot(loss,label='Training Set Loss')
plt.plot(validation_loss,label='validation loss')
plt.ylabel('Loss')
plt.title('Traing and Validation Loss Across Epochs')
plt.legend()
```

```python
confusion_matrix=metrics.confusion_matrix
print('confusion matrix')
c_m=confusion_matrix(y_test,y_pred)
print(c_m)
```

```python
result=metrics.classification_report(y_test,y_pred)
print('result',result)
```

```python
# or we can use the seaborn heatmap to visualize the confusion matrix:
ax=sns.heatmap(c_m,cmap='flare',annot=True,fmt='d')
plt.xlabel('Predicted Class',fontsize=12)
plt.ylabel('True Class',fontsize=12)
plt.title('Confusion Matrix',fontsize=12)
plt.show()
```
