## Let' s have another test to check if the data is normaly distributed:
#### I could use shapiro.test but the data points are more then 5000 so this kind of test will be no accurate
#### I can use instead: Kolmogorov-Smirnov Test:

```python
from scipy.stats import kstest

# Run the Kolmogorov-Smirnov test
res_for_normality = kstest(data['Arrival Delay'], 'norm')

#Check the p-value
if res_for_normality.pvalue > 0.05:
    print("Data appears to be normally distributed")
else:
    print("Data does not appear to be normally distributed")
```

# Gridsearch

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Define a parameter grid for hyperparameter tuning
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

# Create a DecisionTreeClassifier instance
decision_tree = DecisionTreeClassifier(random_state=42)

# Create a GridSearchCV instance
grid_search = GridSearchCV(decision_tree, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to your data (X, y, or your specific data)
grid_search.fit(X_train, y_train)

# Get the best parameters and best Decision Tree model
best_params = grid_search.best_params_
best_decision_tree = grid_search.best_estimator_

print("Best Parameters:", best_params)
print("Best Decision Tree Model:", best_decision_tree)
```

# create and train the model:


```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=20,
                                  min_samples_leaf=2,min_samples_split=10,splitter='random')
classifier.fit(X_train,y_train)   
```

# predict y_pred based on X_train so that later you can evaluate the model comparing y_pred to y_true(y_test)

```python
y_pred=classifier.predict(X_test)
y_pred
```

# create a dataframe to compare the 2 variables:

```python
comparison=pd.DataFrame(y_test,columns=['y_true'])
comparison['y_perd']=y_pred
comparison
```

# evluate the model:

```python
from sklearn import metrics
accur=metrics.accuracy_score(y_test,y_pred)
print('accuracy is: ',accur)
```

```python
confusion_matrix=metrics.confusion_matrix
print('confusion matrix')
c_m=confusion_matrix(y_test,y_pred)
print(c_m)
```

```python
result=metrics.classification_report(y_test,y_pred)
print('result',result)
```

```python
# or we can use the seaborn heatmap to visualize the confusion matrix:
ax=sns.heatmap(c_m,cmap='flare',annot=True,fmt='d')
plt.xlabel('Predicted Class',fontsize=12)
plt.ylabel('True Class',fontsize=12)
plt.title('Confusion Matrix',fontsize=12)
plt.show()
```

## Evaluate the model using  Cross Validation:


```python
from sklearn.model_selection import cross_val_score
cross_val_result=cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10)
print(' the cross validation result is ',np.mean(cross_val_result),'with a standard deviation between scores  of: ',np.std(cross_val_result))
```
