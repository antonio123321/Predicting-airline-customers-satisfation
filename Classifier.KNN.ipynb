## Let' s have another test to check if the data is normaly distributed:
#### I could use shapiro.test but the data points are more then 5000 so this kind of test will be no accurate
#### I can use instead: Kolmogorov-Smirnov Test:

```python
from scipy.stats import kstest

# Run the Kolmogorov-Smirnov test
res_for_normality = kstest(data['Arrival Delay'], 'norm')

#Check the p-value
if res_for_normality.pvalue > 0.05:
    print("Data appears to be normally distributed")
else:
    print("Data does not appear to be normally distributed")
```

```python
# scale all data which is not 0 or 1
# I could avoid this step as decision tree is robust to unscaled data


from sklearn.preprocessing import MinMaxScaler      #I use min-max scaler so all of the values will be between 1 and  0

sc=MinMaxScaler()
X_train[:,4:]=sc.fit_transform(X_train[:,4:]) # here you need to scale all the column with numerical values that are not 0 or one's. meaning from Age(index 4) forward
X_test[:,4:]=sc.transform(X_test[:,4:])
X_train[0]
```

## Check for the optimal number of k:

```python
from sklearn.datasets import make_classification

from sklearn.neighbors import KNeighborsClassifier

# Define a range of k values to test
k_values = range(1, 20)

# Initialize empty lists to store accuracy scores
accuracy_scores = []

# Iterate through different values of k
for k in k_values:
    # Create and fit the k-NN classifier
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Calculate the accuracy on the test set
    accuracy = knn.score(X_test, y_test)
    accuracy_scores.append(accuracy)

# Plot the accuracy scores for different k values
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores)
plt.title('k-NN Accuracy vs. Number of Neighbors (k)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.grid(True)

```

# apply search grid:

```python

from sklearn.model_selection import GridSearchCV

# Define a parameter grid for hyperparameter tuning
param_grid = {
    'n_neighbors': [3, 5, 7, 9,11,13,15],  # Number of neighbors
    'weights': ['uniform', 'distance'],  # Weighting method
    'p': [1, 2],  # Distance metric (1 for Manhattan, 2 for Euclidean)
}

# Create a KNeighborsClassifier instance
knn = KNeighborsClassifier()

# Create a GridSearchCV instance
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to your data (X_train, y_train, or your specific data)
grid_search.fit(X_train, y_train)

# Get the best parameters and best KNN model
best_params = grid_search.best_params_
best_knn = grid_search.best_estimator_

print("Best Parameters:", best_params)
print("Best KNN Model:", best_knn)
```

# create and train the model:


```python
from sklearn.neighbors import KNeighborsClassifier  # p shows the power parameter for the Minkowski metric.
classifier=KNeighborsClassifier(n_neighbors=9,metric='minkowski',p=1,weights='distance') #When p = 2, this is equivalent to euclidean_distance. 
classifier.fit(X_train,y_train)              
```

# predict y_pred based on X_train so that later you can evaluate the model comparing y_pred to y_true(y_test)

```python
y_pred=classifier.predict(X_test)
y_pred
```

# create a dataframe to compare the 2 variables:

```python
comparison=pd.DataFrame(y_test,columns=['y_true'])
comparison['y_perd']=y_pred
comparison
```

# evaluate the model:

```python
from sklearn import metrics
accur=metrics.accuracy_score(y_test,y_pred)
print('accuracy is: ',accur)
```

```python
confusion_matrix=metrics.confusion_matrix
print('confusion matrix')
c_m=confusion_matrix(y_test,y_pred)
print(c_m)
```

```python
result=metrics.classification_report(y_test,y_pred)
print('result',result)
```

```python
# or we can use the seaborn heatmap to visualize the confusion matrix:
ax=sns.heatmap(c_m,cmap='flare',annot=True,fmt='d')
plt.xlabel('Predicted Class',fontsize=12)
plt.ylabel('True Class',fontsize=12)
plt.title('Confusion Matrix',fontsize=12)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_result=cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10)
print(' the cross validation result is ',np.mean(cross_val_result),'with a standard deviation between scores  of: ',np.std(cross_val_result))
```
